```python
# Importamos las librerías básicas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**TODOS LOS RIESGOS PUEDEN SER REPRESENTADOS MEDIANTE LA PROBABILIDAD E IMPACTO.**

- **PROBABILIDAD:** nos indica que tan probable es que ocurra un riesgo formando un criterio de probabilidad. 

NOTA: El riesgo puede ser de poco impacto, pero si su probabilidad es alta puede representar una pérdida grave a futuro.


- **IMPACTO/SEVERIDAD:** Nos indica el criterio de gravedad o qué tanto afecta a la organización, esto puede ser en términos monetarios. 

NOTA: Si el riesgo es de muy poca probabilidad pero de un gran impacto puede generar un evento catastrófico.

```python
# Denominamos algunos riesgos en términos de su probabilidad e impacto

riesgos = {
'Riesgo': ['Pérdida de datos', 'Riesgo regulatorio', 'Riesgo de seguridad', 'Riesgo de integridad'],
'Impacto': [3, 2, 1, 9],              # Impacto entre el 1 y 10
'Probabilidad': [0.33, 0.4, 0.2, 0.1],    # Probabilidad de ocurrencia
}

df = pd.DataFrame(riesgos)
```

```python
# Calculamos el nivel de gravedad del riesgo
df['Nivel de Riesgo'] = df['Impacto'] * df['Probabilidad'] 

# Mostramos los datos
print(df)
```

```python
import seaborn as sns

# Configuración de estilo
sns.set_style("whitegrid")
plt.figure(figsize=(12, 8))
colors = sns.color_palette("coolwarm", len(df['Riesgo']))
bars = plt.bar(df['Riesgo'], df['Nivel de Riesgo'], color=colors, edgecolor='black')

# Etiquetas y título
plt.xlabel('Riesgo', fontsize=14, labelpad=10)
plt.ylabel('Nivel de Riesgo', fontsize=14, labelpad=10)
plt.title('ANÁLISIS DE RIESGOS', fontsize=16, pad=15)

# Rotar etiquetas del eje x si es necesario
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()
```
